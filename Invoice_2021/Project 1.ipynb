```python

summary_stats = df.describe()
# Plotting the summary statistics
fig, ax = plt.subplots()
summary_stats.plot(kind='bar', ax=ax)
ax.set_xticklabels(summary_stats.index, rotation=0)
plt.ylabel('Value')
plt.title('Summary Statistics')
plt.tight_layout()
plt.show()
```

```python
summary_stats=summary_stats.transpose()
for column in summary_stats.columns:
    fig, ax = plt.subplots()
    summary_stats[column].plot(kind='bar', ax=ax)
    ax.set_xticklabels(summary_stats.index, rotation=0)
    plt.ylabel('Value')
    plt.title(f'Summary Statistics - {column}')
    plt.tight_layout()
    plt.show()
```
